```python
hash_letter = {' ': 0,  '!': 1, "'": 2, ',': 3, '?': 4, 'a': 5, 'b': 6, 'c': 7,
               'd': 8, 'e': 9, 'f': 10, 'g': 11, 'h': 12, 'i': 13, 'j': 14,
               'k': 15, 'l': 16, 'm': 17, 'n': 18, 'o': 19, 'p': 20, 'q': 21,
               'r': 22, 's': 23, 't': 24, 'u': 25, 'v': 26, 'w': 27, 'x': 28,
               'y': 29}

hash_numbers = {0:' ', 1:'!', 2:"'", 3:',', 4:'?', 5:'a', 6:'b', 7:'c', 8:'d', 
                9:'e', 10:'f', 11:'g', 12:'h', 13:'i', 14:'j', 15:'k', 16:'l', 
                17:'m', 18:'n', 19:'o', 20:'p', 21:'q', 22:'r', 23:'s', 24:'t', 
                25:'u', 26:'v', 27:'w', 28:'x', 29:'y'}
```

```python
import numpy as np

def generateLyric2(model, first, max):
  res = 'Lyric: ' + first

  pattern = [hash_letter[first]]
  for i in range(max):
    x = np.reshape(pattern, (1, len(pattern), 1))
    x = x / float(30)

    y = model.predict(x, verbose=0)
    y = np.argmax(y)
    pattern.append(y)
    pattern = pattern[1:len(pattern)]

    res = res + hash_numbers[y]
  return res
```

```python
generateLyric2(model1, 'n', 500)
```

```python
generateLyric2(model2, 'n', 500)
```
