```python
import pandas as pd
import re
from unidecode import unidecode

def getType(name):
    try:
        # format the name
        name = re.sub("[\(\[].*?[\)\]]", "", name)
        name = name.rstrip()
        name = unidecode(name)
        name = name.replace(" ", "_")
        # search table
        table = pd.read_html('https://en.wikipedia.org/wiki/' + name)
        # get the data
        df = table[0]
        data = list(df[ df[0] == 'Type' ][1])[0]
        data = data.split()[0]
    except:
        data = 'NA'
    return data

#getType('Universidad Nacional Autónoma de México (UNAM)')
ds['type'] = ds.apply(lambda x: getType(x['institution']), axis=1)
```

```python
import pandas as pd
import re
from unidecode import unidecode
import types

def getIndex(table):
    i = 0
    while i < len(table):
        df = table[i]
        if df[ df[0] == 'Type' ].empty:
            i = i + 1
        else:
            break
    return i;

def getType(name, utype):
    if utype in ('National', 'National,', 'Nonprofit', 'Grande', 'French', 'public', 'Autonomous', 
                 'Independent/Partly', 'State', 'State-supported', 'Central', 'State', 'state',
                 'Technical', 'Governmental', 'Classical', 'Liberal'):
        return 'Public'
    
    if 'Public' in utype:
        return 'Public'
    
    if utype in ('Prevailed', 'Tradional,', 'Royal', 'Pontifical,', 'Deemed', 'Joint-Stock', 'Traditional'):
        return 'Private'
    
    if 'Private' in utype:
        return 'Private'
    
    if utype != '-':
        return utype
    
    if "|" in name:
        name = name.split('|')
        name = name[0]
    
    if "," in name:
        tmp = name.split(',')
        if len(tmp[0]) > len(tmp[1]):
            name = tmp[0]
        else:
            name = tmp[1]
    
    if "-" in name:
        tmp = name.split('-')
        if len(tmp[0]) > len(tmp[1]):
            name = tmp[0]
        else:
            name = tmp[1]
    
    # format the name
    name = re.sub("[\(\[].*?[\)\]]", "", name)
    name = name.rstrip()
    name = unidecode(name)
    name = name.replace(" ", "_")
    
    try:
        # search table
        #print('https://en.wikipedia.org/wiki/' + name)
        table = pd.read_html('https://en.wikipedia.org/wiki/' + name)
        # get the data
        i = getIndex(table)
        df = table[i]
        data = list(df[ df[0] == 'Type' ][1])[0]
        data = data.split()[0]
    except:
        data = '-'
    return data
```
